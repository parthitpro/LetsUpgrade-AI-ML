## Question 1:Using the given MT cars dataset you need to find outlier with z score.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Users\\raghav\\Desktop\\assignments\\day-17\\mtcars.csv")
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.columns
```

```python
data = [df.mpg,df.cyl,df.disp,df.hp,df.drat,df.wt,df.qsec,df.vs,df.am,df.gear,df.carb]
fig = plt.figure(figsize =(10,10))
bp = plt.boxplot(data,vert=False,labels=(['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am',
       'gear', 'carb'])) 
plt.show()
```

From the Above Box plot we can confirm that the "hp" has  Outliers.

```python
outlier = []
for i in df['hp']:
    z = (i-df['hp'].mean())/df['hp'].std()
    if np.abs(z) > 2:
        outlier.append(i)
print(outlier)
```

From the Boxplot we were able to detect the "hp" has an outlier.Further We found the value which is an Outlier.

## Question 2: Using the Boston data set from sklearn, find out the Outlier and treat them.

```python
from sklearn.datasets import load_boston

data = load_boston()
df1 = pd.DataFrame(data.data, columns=data.feature_names)
df1.head()
```

```python
df1.columns
```

```python
bdata = [df1.CRIM,df1.ZN,df1.INDUS,df1.CHAS,df1.NOX,df1.RM,df1.AGE,df1.DIS,df1.RAD,df1.TAX,df1.PTRATIO,df1.B,df1.LSTAT]
fig = plt.figure(figsize =(20,10))
bD = plt.boxplot(bdata,vert=False,labels=(['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT'])) 
plt.show()
```

```python
from scipy import stats
zscore = np.abs(stats.zscore(df1))
print(zscore)
```

```python
thresold = 3
print(np.where(zscore >3))
```

* The above are the loactions where Outliers exists,its in the form of rows & columns.
* This shall be cleaned as below:

```python
clean_boston = df1[(zscore < 3).all(axis=1)]
```

```python
clean_boston.shape
```

As we can see the 91 rows with outliers have been removed.
